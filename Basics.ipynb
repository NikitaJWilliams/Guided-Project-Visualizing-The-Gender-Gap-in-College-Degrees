# Guided Project: Visualizing The Gender Gap in College Degrees

## Introduction

In this project, I will explore the gender gap across college majors with the help of data visualization tools. The American Department of Education compiles and publishes the data set containing the percentage of bachelor's degrees granted to women from 1970 to 2012. Each column represents a seperate degree type. Randal Olson, a data scientist at University of Pennsylvania, has cleaned the data set and made it publicly available on his [webpage](http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv). I will be using his version of the data set to determine whether the gender gap actually exists in the education industry.

## Data Analysis

Let's go ahead and visualize the percentage females and males graduating with American STEM degrees across time.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

The above line charts suggest that the US STEM graduates in the early 1970s were predominantly male; however, in the mid 1970s, all that changed as the percentage of women successfully completing psychology majors soared just over 50%. Similarly, in the late 1980s, the gender gap reversed within biology. On the other hand, men continued to dominate engineering and computer science programs from 1970s all the way up to 2011.

## Gender Gap Comparison Across All Disciplines

```python
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig_two = plt.figure(figsize = (16,20))
for l in range(5):
    ax2 = fig_two.add_subplot(4,3,l+1)
    ax2.plot(women_degrees['Year'], women_degrees[lib_arts_cats[l]],label = "Women", c = cb_dark_blue, linewidth = 3)
    ax2.plot(women_degrees['Year'], 100 - women_degrees[lib_arts_cats[l]],label = "Men", c = cb_orange, linewidth = 3)
    ax2.spines["right"].set_visible(False)
    ax2.spines["left"].set_visible(False)
    ax2.spines["top"].set_visible(False)
    ax2.spines["bottom"].set_visible(False)
    ax2.set_ylim(0,100)
    ax2.set_xlim(1968,2011)
    ax2.set_title(lib_arts_cats[l])
    ax2.set_yticks([0,100])
    #hide horizontal ticks
    ax2.tick_params(bottom = "off", top = "off", left = "off", right = "off",labelbottom = "off")
    if l == 0:
        ax2.text(2005, 87, 'Women')
        ax2.text(2002, 8, 'Men')
    elif l == 4:
        ax2.text(2005, 62, 'Men')
        ax2.text(2001, 35, 'Women') 
    ax2.axhline(50, c = (171/255,171/255,171/255), alpha = 0.3)
for o in range(5,11):
    ax3 = fig_two.add_subplot(4,3,o+1)
    ax3.plot(women_degrees['Year'], women_degrees[other_cats[o-5]],label = "Women", c = cb_dark_blue, linewidth = 3)
    ax3.plot(women_degrees['Year'], 100 - women_degrees[other_cats[o-5]],label = "Men", c = cb_orange, linewidth = 3)
    ax3.spines["right"].set_visible(False)
    ax3.spines["left"].set_visible(False)
    ax3.spines["top"].set_visible(False)
    ax3.spines["bottom"].set_visible(False)
    ax3.set_ylim(0,100)
    ax3.set_xlim(1968,2011) 
    ax3.set_title(other_cats[o-5])
    ax3.set_yticks([0,100])
    ax3.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    if o == 5:
        ax3.text(2005, 87, 'Women')
        ax3.text(2002, 8, 'Men')
    elif o == 10:
        ax3.text(2005, 62, 'Men')
        ax3.text(2001, 35, 'Women') 
    #hide horizontal ticks    
    if o < 8:
        ax3.tick_params(bottom = "off", top = "off", left = "off", right = "off",labelbottom = "off")
```

From 1968 to 2011, women dominated the liberal arts space with the exception of social science and history. They also made up more than half of public administration, health and education graduates. Agriculture, business and architecture majors used to be male domains up until 2000s when the number of male and female degree holders converged to roughly 50%.
